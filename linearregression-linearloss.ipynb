```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np
import random
from matplotlib import pyplot as plt
```

```python
#Smaller power loss fucntion performs well with fast learning rate and decent no of epochs whereas larger power loss function needs very small learning rate otherwise overflow will take place
n = 1 #power of loss function
size=200  #size of dataset
epochs=1000
learning_rate=0.005
```

```python
#Generating a random data structure

x=5*(np.random.rand(size))+2  #Creates a array of (size) random numbers with mean 5 and Standard deviation 2
extra=0.5*(np.random.rand(size))
y_real=(10*x)+2+extra
# print(x)
# print(y_real)
```

```python
w1=np.random.rand(1)
w2=0

def compare(n):
    if(n>0):
        return 1
    else:
        return -1
    
for i in range(epochs):
    y_pred=w1*x+w2
    loss=pow(abs(y_pred-y_real),n)
    cost=(np.sum(loss))/size
#     print("loss :" + str(cost))

    #Backtracking
    dw1=(np.dot(n*(pow(abs(y_real-y_pred),(n-1)))*list(map(compare,(y_pred-y_real))),x))/size
#     print(dw1)
    dw2=(np.sum(n*(pow(abs(y_real-y_pred),(n-1)))*list(map(compare,(y_pred-y_real)))))/size
#     print(dw2)

    w1=w1-(learning_rate*dw1)
#     print(w1)
    w2=w2-(learning_rate*dw2)
#     print(w2)
print("parameters :")
print("w1 : %f"%w1)
print("w2 : %f"%w2)
```

```python
#testing the model on trained parameters
print(n)
test_x=np.random.rand(100)
y_actual_test=(10*test_x)+2
y_pred_test=w1*test_x+w2

#plotting the result
plt.figure(figsize=(12, 6))
plt.plot(test_x,y_pred_test)
plt.plot(test_x,y_actual_test,'ro')
plt.title("Actual vs Predicted")
plt.xlabel('InputX')
plt.ylabel('OutputY')

plt.show()
```
