```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np
import random
from matplotlib import pyplot as plt
```

```python
n = 4 #power of loss function
size=100  #size of dataset
epochs=1000000
learning_rate=1e-9
```

```python
#Generating a random data structure

x=5*(np.random.rand(size))+2  #Creates a array of (size) random numbers with mean 5 and Standard deviation 2
extra=0.5*(np.random.rand(size))  #Noise in the input
y_real=(10*x*x)+(8*x)+9+extra
# print(x)
# print(y_real)
```

```python
w1=np.random.rand(1)
w2=np.random.rand(1)
w3=0

def compare(n):
    if(n>0):
        return 1
    else:
        return -1
    
for i in range(epochs):
    y_pred=(w1*x*x)+(w2*x)+w3
    loss=pow(abs(y_pred-y_real),n)
    cost=(np.sum(loss))/size
#     print("loss :" + str(cost))

    #Backtracking
    
    dw1=(np.dot(n*(pow(abs(y_real-y_pred),(n-1)))*list(map(compare,(y_pred-y_real))),x*x))/size
#     print(dw1)
    dw2=(np.dot(n*(pow(abs(y_real-y_pred),(n-1)))*list(map(compare,(y_pred-y_real))),x))/size
#     print(dw2)
    dw3=(np.sum(n*(pow(abs(y_real-y_pred),(n-1)))*list(map(compare,(y_pred-y_real)))))/size
#     print(dw3)

    w1=w1-(learning_rate*dw1)
#     print(w1)
    w2=w2-(learning_rate*dw2)
#     print(w2)
    w3=w3-(learning_rate*dw3)
#     print(w3)
print("Trained parameters :")
print("w1 : %f"%w1)
print("w2 : %f"%w2)
print("w3 : %f"%w3)
```

```python
#testing the model on trained parameters
print(n)
test_x=np.random.rand(100)
y_actual_test=(10*test_x*test_x)+(6*test_x)+2
y_pred_test=(w1*test_x*test_x)+(w2*test_x)+w3

#plotting the results
plt.figure(figsize=(12, 6))
plt.plot(test_x,y_pred_test)
plt.plot(test_x,y_actual_test,'ro')
plt.title("Actual vs Predicted")
plt.xlabel('InputX')
plt.ylabel('OutputY')

plt.show()
```
